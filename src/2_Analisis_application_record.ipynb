## Análisis de Datos application record

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df_ant=pd.read_csv('../archive/application_credit_record.csv')
df = df_ant.drop(columns=["Unnamed: 0","CODE_GENDER","NAME_INCOME_TYPE","NAME_EDUCATION_TYPE","NAME_FAMILY_STATUS","FLAG_MOBIL","FLAG_WORK_PHONE","FLAG_EMAIL","OCCUPATION_TYPE","CNT_CHILDREN","CNT_FAM_MEMBERS","FLAG_PHONE","AGE_YEARS"])
df.head()
```

```python
df.describe()
```

```python
#Descripción de FLAG_OWN_CAR
column = 'FLAG_OWN_CAR'
valores = df[column].value_counts()
print(f'Valores de {column}: {valores}')
plt.figure(figsize=(10,6))
plt.bar(valores.index,valores.values)
plt.title(f'Valores {column}')
plt.xlabel('Banderas')
plt.ylabel('Cantidad de personas')
plt.xticks([0, 1], ['0', '1'])
valores_car = valores
```

```python
#Descripción de FLAG_OWN_REALTY
column = 'FLAG_OWN_REALTY'
valores = df[column].value_counts()
print(f'Valores de {column}: {valores}')
plt.figure(figsize=(10,6))
plt.bar(valores.index,valores.values)
plt.title(f'Valores {column}')
plt.xlabel('Banderas')
plt.ylabel('Cantidad de personas')
plt.xticks([0, 1], ['0', '1'])
valores_propiedad = valores
```

```python
plt.figure(figsize=(5,6))
plt.scatter(valores_car.index,valores_car.values,label='Auto',color='green')
plt.scatter(valores_propiedad.index,valores_propiedad.values,label='Propiedad', color='blue')
plt.title(f'Valores FLAG_OWN_REALTY y FLAG_OWN_CAR')
plt.xlabel('Banderas')
plt.ylabel('Cantidad de personas')
plt.legend()
plt.xticks([0, 1], ['0', '1'])
```

```python
#Descripción de NAME_HOUSING_TYPE
column = 'NAME_HOUSING_TYPE'
valores = df[column].value_counts()
print(f'Valores de {column}: {valores}')
plt.figure(figsize=(20,10))
plt.bar(valores.index,valores.values)
plt.title(f'Valores {column}',fontsize=20)
plt.xlabel('Categorías',fontsize=20)
plt.ylabel('Cantidad de personas',fontsize=20)
plt.tick_params(axis='both', labelsize=16)
```

```python
#Descripción de AMT_INCOME_TOTAL
column = 'AMT_INCOME_TOTAL'
mediana = df[column].median()
media = df[column].mean()
q1 = df[column].quantile(0.25)
q2 = df[column].quantile(0.50)
q3 = df[column].quantile(0.75)
print("Primer cuartil (25%):", q1)
print("Segundo cuartil o Mediana (50%):", q2)
print("Tercer cuartil (75%):", q3)
print("Media:", media)
sns.boxplot(x=df[column])
plt.title('Boxplot Ingreso total')
plt.xlabel('Unidades monetarias')
```

```python
#Descripción de MONTHS_EMPLOYED
column = 'MONTHS_EMPLOYED'
mediana = df[column].median()
media = df[column].mean()
q1 = df[column].quantile(0.25)
q2 = df[column].quantile(0.50)
q3 = df[column].quantile(0.75)
print("Primer cuartil (25%):", q1)
print("Segundo cuartil o Mediana (50%):", q2)
print("Tercer cuartil (75%):", q3)
print("Media:", media)
sns.boxplot(x=df[column],color='green')
plt.title('Boxplot Meses empleado')
plt.xlabel('Meses')
```

```python
#Descripción de timespan
column = 'timespan'
mediana = df[column].median()
media = df[column].mean()
q1 = df[column].quantile(0.25)
q2 = df[column].quantile(0.50)
q3 = df[column].quantile(0.75)
print("Primer cuartil (25%):", q1)
print("Segundo cuartil o Mediana (50%):", q2)
print("Tercer cuartil (75%):", q3)
print("Media:", media)
sns.boxplot(x=df[column],color='orange')
plt.title('Boxplot del tiempo de vida de los créditos')
plt.xlabel('timespan (Meses)')
```
